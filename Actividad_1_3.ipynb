```python
df1 = pd.read_excel('Inmuebles.xlsx', skiprows = 1)
df1
```

```python
df1 = df1.drop(['Referencia'], axis = 1)
filtroa_1_3 = df1
filtroa_1_3
```

```python
filtrob_1_3 = df1[(df1['Fecha_Alta'] == '2007-04-29 00:00:00') | (df1['Fecha_Alta'] == '2004-01-02 00:00:00')]
filtrob_1_3
```

```python
filtroc_1_3 = df1[(df1['Tipo'] == 'Parking') | (df1['Tipo'] == 'Local') | (df1['Tipo'] == 'Industrial') | (df1['Tipo'] == 'Piso')]
filtroc_1_3
```

```python
filtrod_1_3 = df1[(df1['Provincia'] == 'Girona') & (df1['Superficie'] > 100)]
filtrod_1_3
```

```python
filtroe_1_3 = df1[(df1['Precio Venta'] > 1000000) | (df1['Vendedor'] == 'Carmen') | (df1['Vendedor'] == 'Joaquín')]
filtroe_1_3
```

```python
filtrof_1_3 = df1[~(df1['Provincia'] == 'Barcelona')]
filtrof_1_3
```

```python
filtrog_1_3 = df1[~(df1['Operación'] == 'Alquiler') & ~(df1['Tipo'] == 'Oficina')]
filtrog_1_3
```

```python
filtroa_1_3.to_csv('Filtroa_Act_1_3.csv')
filtrob_1_3.to_csv('Filtrob_Act_1_3.csv')
filtroc_1_3.to_csv('Filtroc_Act_1_3.csv')
filtrod_1_3.to_csv('Filtrod_Act_1_3.csv')
filtroe_1_3.to_csv('Filtroe_Act_1_3.csv')
filtrof_1_3.to_csv('Filtrof_Act_1_3.csv')
filtrog_1_3.to_csv('Filtrog_Act_1_3.csv')
```
