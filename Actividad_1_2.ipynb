```python
df.CVE_CLPV = df.CVE_CLPV.replace({'O': '0',
                                   '6781|': '6781',
                                   '3612C': '3612',
                                   '      ': '0',
                                   '2385A': '2385',
                                   '2510A': '2510',
                                   '8440O': '8440',
                                   '8440Q': '8440',
                                   'M' : '0'}, regex = False)

df['CVE_CLPV'] = df['CVE_CLPV'].astype(float)
```

```python
filtro1_1_2 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1_1_2
```

```python
filtro2_1_2 = df[~(df['CVE_VEND'] == 5) & ~(df['CVE_VEND'] == 4)]
filtro2_1_2
```

```python
filtro2_1_2.groupby(['CVE_VEND'])['CVE_VEND'].count()
```

```python
df['FECHA_ENT'] = df['FECHA_ENT'].astype(str)
```

```python
filtro3_1_2 = df[df['FECHA_ENT'] == "2022-04-13 00:00:00"]
filtro3_1_2
```

```python
filtro4_1_2 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]
filtro4_1_2
```

```python
filtro5_1_2 = df.iloc[ : , [0, 6, 7, 9]]
filtro5_1_2
```

```python
filtro6_1_2 = df.iloc[7001:7100,:]
filtro6_1_2
```
